```python
# Célula 01 - Carregar o dataset Iris e visualizar as primeiras linhas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("iris")
df.to_csv("iris.csv")
```

```python
# Célula 02 - Informações básicas sobre o Dataframe (info)
df.info()
```

```python
# Célula 03 - Estatísticas descritivas básicas. Dica: É um método do dataframe
display(df.describe(include='all').T)
```

```python
# Célula 04 - Contagem de quantas flores existem por espécie
df['species'].value_counts()
```

```python
# Célula 05 - Histograma do campo sepal_length
plt.figure(figsize=(6, 4))

# plt.hist(df['sepal_length'], bins=20)
sns.histplot(df['sepal_length'], bins=20, kde=True)
plt.title("Distribuição de sepal_length")
plt.xlabel('sepal_length')
plt.ylabel("Frequência")
plt.show()
```

```python
# Célula 06 - Boxplot de petal_length por espécie
plt.figure(figsize=(6, 4))

sns.boxplot(data=df, x='species', y='petal_length')
plt.title("petal_length por espécie")
plt.xlabel("Espécie")
plt.ylabel("petal_length")

plt.show()
```

```python
# Célula 07 - Relação entre sepal_length e sepal_width
# Nesse caso iremos utilizar o gráfico scatterplot
plt.figure(figsize=(8, 4))

sns.scatterplot(
    data=df,
    x='sepal_length',
    y='sepal_width',
    hue='species'
)

plt.title("Relação entre sepal_length e sepal_width")
plt.show()
```

```python
# Célula 08 - Calcule a média de sepal_width para cada espécie
df.groupby('species')['sepal_width'].mean()
```

```python
# Célula 09 - Calcular quantas flores da espécie *setosa* têm petal_length menor que 1.5

# Dentro dos colchetes do dataframe podemos colocar uma operação com operadores booleanos.
# Será retornado um novo dataframe para cada linha que retorne 'True' para a condição.
setosa = df[df['species'] == 'setosa']
setosa_pequena = setosa[setosa['petal_length'] < 1.5]

print(f"Quantidade de flores da espécie 'setosa' com o comprimento da pétala menor que 1.5cm: {len(setosa_pequena)}")
```
