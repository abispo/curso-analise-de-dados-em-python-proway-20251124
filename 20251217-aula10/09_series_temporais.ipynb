# Séries temporais

Séries temporais são dados coletados ao longo do tempo, em ordem cronológica. Ou seja, o tempo importa.

## Exemplos reais

* Vendas diárias
* Preço de ações
* Temperatura ao longo do tempo
* Consumo mensal de energia

## Componentes de uma série temporal

* Tendência         -> Crescimento ao longo em tempo
* Sazonalidade      -> Padrões que se repetem
* Ruído             -> Variação aleatória

```python
# Analizando a evolução de vendas ao longo de 12 meses

import pandas as pd
import matplotlib.pyplot as plt

# Criação do índice temporal (datas mensais)
datas = pd.date_range(start="2024-01-01", periods=12, freq="M")

# Valores simulados das vendas
vendas = [120, 130, 125, 140, 150, 160, 170, 175, 155, 160, 170, 180]

df = pd.DataFrame({
    "data": datas,
    "vendas": vendas
})

# Definimos um índice
df.set_index("data", inplace=True)
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['vendas'], marker='o')
plt.title("Série Temporal de Vendas")
plt.xlabel("Tempo")
plt.ylabel("Vendas")
plt.show()
```
