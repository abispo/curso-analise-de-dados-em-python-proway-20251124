```python
# Medidas de centralidade

# Úteis para responder a seguinte pergunta: "Qual é um valor representativo dos meus dados?"

import pandas as pd
df = pd.read_csv("wine_quality_red.csv", delimiter=';')

col = 'alcohol'

# Média
# Soma dividida pelo número de valores. Utilizada quando os dados são simétricos e quando não existem muitos outliers. Um exemplo válido de uso da média é quando queremos calcular a idade média dos frequentadores de um shopping
print(f"Média de graduação alcoólica: {df[col].mean():.2f}")

# Mediana
# É o valor central da nossa população/amostra (50%). Diferente da média, a mediana é resistente a outliers e é mais representativa em distribuições assimétricas.
# Exemplo do mundo real: Imagine os seguintes salários de uma empresa:
# Funcionario A: 2000
# Funcionario B: 2500
# Funcionario C: 3000
# Diretor: 50000
# Média 14375.
# Mediana: 2750
print(f"Mediana da graduação alcóolica: {df[col].median():.2f}.")

# Moda
# É o valor mais frequente de uma distribuição. Geralmente utilizada para responder perguntas do tipo "Qual o produto mais vendido"
# Exemplo: "Qual o combo mais vendido na lanchonete tal?"
print(f"Moda da graduação alcóolica: {df[col].mode()[0]:.2f}")
```
